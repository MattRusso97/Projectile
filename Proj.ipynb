Let's import the data and the needed libraries




```python
%matplotlib notebook
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

VxA, VyA = np.loadtxt("analitical.txt", unpack = True, usecols = (0, 1))
VxR, VyR = np.loadtxt("runge.txt", unpack = True, usecols = (0, 1))

#Time array
t = np.linspace(0, 100, num = 1000)
```

Let's plot and compare the two integration methods

```python
fig = plt.figure(figsize = (15, 8))
ax1 = fig.add_subplot(121, xlabel = "time [s]", ylabel = "v(t) [m/s]", title = r"x component of the velocity in function of time - $v_x (t)$", xlim = [0, 100], ylim = [14, 15.2])
ax1.plot(t, VxA, color = "black", label = r"${v}_{x} (t) - Analitical Method$")
ax1.plot(t, VxR, color = "red", linestyle = ":", label = r"${v}_{x} (t) - RungeKutta Method$")
ax1.tick_params(direction = "in", bottom = True, top = True, left = True, right = True)
ax1.grid()
plt.legend()

ax2 = fig.add_subplot(122, xlabel = "time [s]", ylabel = "v(t) [m/s]", title = r"y component of the velocity in function of time - $v_y (t)$", xlim = [0, 2], ylim = [12.5, 15.5])
ax2.plot(t, VyA, color = "black", label = r"${v}_{y} (t) - Analitical Method$")
ax2.plot(t, VyR, color = "red", linestyle = ":", label = r"${v}_{y} (t) - RungeKutta Method$")
ax2.tick_params(direction = "in", bottom = True, top = True, left = True, right = True)
ax2.grid()
plt.legend()

plt.show()
```
